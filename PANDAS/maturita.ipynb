```python
u202 = pandas.read_csv("u202.csv")
u203 = pandas.read_csv("u203.csv")
u302 = pandas.read_csv("u302.csv")
u202
```

```python
# kterým studentům nechybí známka ve třídě 202

# pomocí negace (vlnky)
# u202[~u202["znamka"].isna()]

# pomocí metody dropna
u202.dropna(subset=["znamka"])
```

```python
u202["místnost"] = "u202"
u203["místnost"] = "u203"
u302["místnost"] = "u302"
```

```python
maturita.to_csv('maturita.csv', index=False)
```
